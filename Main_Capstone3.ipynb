## Metric Evaluation

Dalam evaluasi kinerja model, digunakan tiga metrik utama, yaitu:

**1. Root Mean Squared Error (RMSE)**
- **Fungsi:** Mengukur seberapa besar kesalahan yang dihasilkan oleh model.
- **Karakteristik:** Memberikan nilai absolut.
- **Perhitungan:** Merupakan pengembangan dari Mean Squared Error (MSE) dengan mengambil akar kuadrat dari MSE.

**2. Mean Absolute Error (MAE)**
- **Fungsi:** Mengukur rata-rata kesalahan absolut dari prediksi model.
- **Karakteristik:** Memberikan nilai absolut.
- **Perhitungan:** Menghitung rata-rata dari selisih absolut antara prediksi dan nilai sebenarnya.

 **3. Mean Absolute Percentage Error (MAPE)**
- **Fungsi:** Mengukur persentase kesalahan absolut rata-rata antara prediksi dan nilai sebenarnya.
- **Karakteristik:** Memberikan nilai relatif.
- **Perhitungan:** Menghitung persentase kesalahan absolut rata-rata antara prediksi dan nilai sebenarnya.

### VIF (Variance Inflation Factor) 

`VIF`, singkatan dari `Variance Inflation Factor`, adalah ukuran statistik yang digunakan untuk menilai multikolinearitas antar variabel prediktor dalam model regresi. Multikolinearitas muncul ketika dua atau lebih variabel prediktor menunjukkan korelasi yang kuat satu sama lain, sehingga menimbulkan tantangan yang signifikan terhadap proses analisis regresi. bersarkan paper [ Coupling fuzzy clustering](https://www.researchgate.net/publication/340104220_Coupling_fuzzy_clustering_and_cellular_automata_based_on_local_maxima_of_development_potential_to_model_urban_emergence_and_expansion_in_economic_development_zones) bahwa jika nilai VIF lebih dari 10 maka akan terjadi multikolinearitas dan harus di drop. Lalu pada artikel [Multicollinearity in Regression Analysis: Problems, Detection, and Solutions](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5710679/) juga menyebutkan bahwa jika nilai VIF lebih dari 10 maka akan terjadi multikolinearitas dan harus di drop. Berdasarkan penjelasan diatas maka akan dilakukan pengecekan VIF untuk mengetahui apakah ada fitur yang berkorelasi tinggi dengan fitur `median_house_value`. Jika ada maka fitur tersebut akan dihapus. Dan yang terakhir pada artikel [Coporatefinance.com](https://corporatefinanceinstitute.com/resources/data-science/variance-inflation-factor-vif/#:~:text=Generally%2C%20a%20VIF%20above%204,that%20needs%20to%20be%20corrected.) menyatakan bahwa Secara umum, VIF di atas 4 atau toleransi di bawah 0,25 menunjukkan kemungkinan adanya multikolinearitas, dan diperlukan penyelidikan lebih lanjut. Jika VIF lebih tinggi dari 10 atau toleransi lebih rendah dari 0,1, terdapat multikolinearitas signifikan yang perlu diperbaiki.


```python
#def calculate_vif(df) only for numeric
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    vif_df = pd.DataFrame(columns = ['Feature', 'VIF'])
    x_var = df.select_dtypes(include=[np.number])
    x_var = x_var.drop(['median_house_value'], axis=1)
    x_var = x_var.dropna()
    vif_df['Feature'] = x_var.columns
    vif_df['VIF'] = [variance_inflation_factor(x_var.values, i) for i in range(len(x_var.columns))]
    return vif_df

calculate_vif(df).sort_values(by='VIF', ascending=False)
```

# Modeling

Pada modeling dilakukan cross validation dengan menggunakan 5 fold. Dengan menggunakan 5 fold maka akan didapatkan 5 nilai RMSE, MAE dan MAPE. Dari 5 nilai tersebut akan diambil nilai rata-rata untuk menentukan nilai RMSE, MAE dan MAPE yang terbaik. Model yang akan diuji yaitu `Linear Regression`, `KNN`, `Decision Tree`, `Random Forest`, `XGBoost` dan `LGBM`. Untuk menentukan model terbaik akan digunakan nilai RMSE, MAE dan MAPE terkecil.

## Cross Validation

```python
#make cross validation
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
#lightgbm
from lightgbm import LGBMRegressor
#xgboost
from xgboost import XGBRegressor
#catboost
from catboost import CatBoostRegressor
#import kfold
from sklearn.model_selection import KFold
#gbr
from sklearn.ensemble import GradientBoostingRegressor

#split data
X = df.drop(['median_house_value'], axis=1)
y = df['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=41)

lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=2024)
rf = RandomForestRegressor(random_state=2024)
xgb = XGBRegressor(random_state=2024)
lgbm = LGBMRegressor(random_state=2024, verbose=-1)
gbr = GradientBoostingRegressor(random_state=2024)

# Pemodelan dalam skala logaritmik, kemudian diinversikan kembali untuk proses interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)
log_lgbm = TransformedTargetRegressor(lgbm, func=np.log, inverse_func=np.exp)
log_gbr = TransformedTargetRegressor(gbr, func=np.log, inverse_func=np.exp)

# List of all models
model = [
    log_lr,
    log_knn,
    log_dt,
    log_rf,
    log_xgb,
    log_lgbm,
    log_gbr
    

]

scaler = RobustScaler(with_centering=False)

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []


for i in model:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=2024)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

### Hyper Parameter Tuning LightGBM


LightGBM menurut [documentation Library LightGBM](https://lightgbm.readthedocs.io/en/stable/) adalah kerangka kerja penguatan gradien yang menggunakan algoritma pembelajaran Decision Tree. Decision Tree yang ditingkatkan oleh gradien merupakan jenis algoritma mesin pembelajaran yang menggunakan pohon keputusan sebagai estimator dari suatu ensemble. Sebuah ensemble terdiri dari banyak model/estimator lemah (pohon keputusan) yang prediksinya digabungkan untuk membuat prediksi akhir.
Ini dirancang untuk didistribusikan dan efisien dengan keunggulan berikut:

- Kecepatan pelatihan yang lebih cepat dan efisiensi yang lebih tinggi.

- Penggunaan memori yang lebih rendah.

- Akurasi yang lebih baik.

- Dukungan untuk pembelajaran paralel, terdistribusi, dan GPU.

- Mampu menangani data dalam skala besar.



Setelah memilih model terbaik dari Modeling, langkah berikutnya adalah mengoptimasi model melalui proses tuning hyperparameter.

Penyetelan hyperparameter, yang juga dikenal sebagai optimisasi model, melibatkan pencarian kombinasi hyperparameter yang paling optimal untuk meningkatkan performa model pembelajaran mesin Anda. Hyperparameter ini adalah parameter yang tidak dapat dipelajari secara langsung oleh model selama proses pelatihan dan harus disesuaikan sebelum pelatihan dimulai. Sebaliknya, parameter-model seperti bobot dalam jaringan saraf atau pembagi dalam pohon keputusan-diatur selama proses pelatihan.









- **Model:** LigthGBM
- **Search Space:** `num_leaves`, `max_depth`, `min_child_samples`, `subsample`, `colsample_bytree`, `learning_rate`, `n_estimators`, `reg_alpha`, `reg_lambda`, `max_bin`, `boosting_type
- **Optimation Criteria:** `MAPE`
- **Search Method:** `RandomizedSearchCV`



#### LightGBM Parameters

| **Parameter**         | **Description**                          | **Default**   |
|-----------------------|------------------------------------------|---------------|
| **boosting_type**     | Variant of boosting                      | gbdt          |
| **learning_rate**     | Speed of model learning                  | 0.1           |
| **max_depth**         | No limit on tree depth                   | -1 (tidak ada batasan) |
| **min_child_samples** | Minimum samples to form a node           | 20            |
| **n_estimators**      | Number of trees                          | 100           |
| **num_leaves**        | Maximum leaves per tree                  | 31            |
| **reg_alpha**         | L1 regularization                        | 0.0           |


```python
#hyper parametertuning log_lgbm
from sklearn.model_selection import RandomizedSearchCV
#import lightgbm
lightgbm = LGBMRegressor(random_state=2024, verbose=-1)
pipe = Pipeline([
    ('preprocessing', transformer),
    # ('scaler', scaler),
    ('model', lightgbm)
])

param_grid = {
    'model__num_leaves': [50, 75, 100],
    'model__max_depth': [10, 20, 30],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__n_estimators': [50, 100, 200],
    'model__min_child_samples': [10, 20, 30],
    'model__reg_alpha': [0.1, 0.5, 1],
    'model__boosting_type': ['gbdt', 'dart', 'goss'],
}

model_cv = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    cv=crossval,
    scoring= 'neg_mean_absolute_percentage_error',
    n_iter=20,
    n_jobs=-1,
    random_state=2024,
)

model_cv.fit(X_train, y_train)
```
